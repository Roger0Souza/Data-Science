__**Lambda Python avançado e compreensões de lista:**__ 

**Lambda é uma forma do python criar funções anônimas, de forma mais simples e curtas.**

```python
my_func = lambda a, b: print(a + b)
my_func(1, 2)

#lambda retorna apenas o resultado de 1 expressão portanto, não é possível ter expressões complexas ou condicionais
#pode executar um método dentro da expressão como print ou input, porem tem-se atentar a usar apenas uma expressão simples
```

```python
#imprimindo um vetor com os números pares de 2 a 100 usando for comum.
my_list = []
for i in range(2, 101, 2):
    my_list.append(i)
print(my_list)
```

```python
# imprimindo usando list comprehension

my_list2 = [i for i in range(2, 101, 2)]
print(my_list2)

#Aqui nota-se ser dispensavel o método append para adicionar elementos a lista, otimizando o tempo de execução e a quantidade de linhas no código
```

```python
lowercase = 'abcdefghijklmnopqrstuvwxyz'
digits = '0123456789'

answer = [letter1 + letter2 + digit1 + digit2 for letter1 in lowercase for letter2 in lowercase for digit1 in digits for digit2 in digits]
print(answer)
```

```python
#criando 2 matrizes multiplicaveis (2x2) usamos @ para fazer a multiplicação de matrizes
#note que o operador * faz a multiplicação elemento por elemento, não a multiplicação de matrizes
a = np.random.normal(0, 1, (4, 3))
b = np.random.normal(0, 1, (3, 2))
#verificando desvio padrão = 1, 
print(np.std(a)) 
print(np.std(b))
'''O desvio padrão que você especifica na função np.random.normal(0, 1, (2,2)) não é o desvio padrão dos valores gerados, mas sim o desvio padrão da distribuição normal que está sendo utilizada para gerar esses valores.

O desvio padrão retornado por np.std(a) pode não ser exatamente 1 porque ele é calculado com base nos valores específicos gerados aleatoriamente na matriz a, não na distribuição normal utilizada para gerar esses valores. A função np.std() calcula o desvio padrão dos valores reais presentes na matriz, que podem variar um pouco da média e do desvio padrão que você especificou na função np.random.normal().

No entanto, à medida que você aumenta o tamanho da amostra (ou seja, a quantidade de números gerados), o desvio padrão dos valores observados tende a se aproximar mais do desvio padrão especificado na distribuição normal que você usou para gerar os números aleatórios. Isso é um resultado da Lei dos Grandes Números.'''

#multiplicação de matrizees
c = a @ b
print(c)
```

```python
# numpy arrays tem métodos interessantes como sum, max, min, mean, std, entre outros
print(c.sum()) #soma todos os elementos da matriz
print(c.max()) #retorna o maior valor
print(c.min()) #retorna o menor valor
print(c.mean()) #retorna a média
print(c.std()) #retorna o desvio padrão
```

__Indexing, Slicing and Iterating__

__Indexing__

```python
#Vemos que a indexação de array unidimensional é exatamente a mesma que de uma lista comum python.
#Para pegar um elemento numa lista unidimensional usamos a indexação simples. EXEMPLO:

a = np.arange(0, 5)[:, np.newaxis ] + np.arange(0, 5)
print(a)
```

```python
print(a[3, 3]) 
print(a[3][2]) 

#nota-se que as duas formas de indexação são equivalentes
```

```python
#Outro exemplo pode-se passar varios indices para uma matriz multidimensional para obter uma lista de elementos.

print(a[[0, 1, 2], [3, 4, 4]])

#aqui retorna uma lista do elementos primeiro 
# linha 0, coluna 3
# linha 1, coluna 4
# linha 2, coluna 4
```

__Boolean Indexing__

Elementos de indexação booleanas permite selecionar elementos arbitrarios de acordo com as condições dadas por operador

```python
#Exemplo:
print(a>5)
#seleciona os elementos maior que 5 de toda matriz e responde com uma matriz de booleanos
#pode-se usar essa matriz para indexar a matriz original e obter apenas os elementos que são maiores que 5
print(a[a>5])
```
