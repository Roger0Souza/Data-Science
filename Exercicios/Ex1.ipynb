 1- Definir uma lista, x,  de 11 números qualquer de numeros inteiros. Transformar a lista x, num dataframe, y, e exibir o boxplot desses dados.

> Bloco com recuo



```python
x = [9, 11, 23, 25, 41, 94, 67, 55, 32, 29, 43]
x_size = np.size(x)
y = pd.DataFrame(x)
print(y)

q1 = np.percentile(x, 25)
q2 = np.percentile(x, 50)
q3 = np.percentile(x, 75)


IQR = q3 - q1  # Intervalo Interquartil (IQR)
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
outliers = [val for val in x if val < lower_bound or val > upper_bound]

plt.boxplot(x, vert=False, showfliers=False, labels=['Dados'])
plt.scatter(outliers, [1] * len(outliers), color='red', marker='o', label='Outliers')
plt.xlabel('Valores')
plt.title('Boxplot dos Dados')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np

file_path = 'iris.data'
iris_data = pd.read_csv(file_path, delimiter=",")

print(iris_data)
```

### [Sobre o Iris ](https://pt.wikipedia.org/wiki/Conjunto_de_dados_flor_Iris)
O conjunto de dados consiste em amostras de três espécies de flores Iris ( Iris setosa, Iris virginica e Iris versicolor). Quatro atributos foram medidos em cada amostra: o comprimento e a largura das sépalas e pétalas, em centímetros.

A estrutura das variáveis no arquivo é:

1. comprimento sépala em cm
2. largura sépala em cm
3. comprimento da pétala em cm
4. largura da pétala em cm
5. classe:
  - Iris Setosa
  - Iris Versicolour
  - Iris Virginica

### 17. Coloque nomes nas colunas do DataFrame de acordo com a descrição acima

```python
columns =  ["Comprimento sépala em cm", "largura sépala em cm", "Comprimento da pétala em cm", "Largura da pétala em cm", "Classe"]

iris_data.columns=columns
print(iris_data)
```

### 18. Usando Pandas, gere um histograma dos valores para cada um dos atributos (colunas)



```python
import matplotlib.pyplot as plt

for column in iris_data.columns:
    iris_data[column].hist(bins=10)
    plt.xlabel('Valores')
    plt.ylabel('Frequência')
    plt.title(f'Histograma das coluna {column}')
    plt.show()
```

### 19. Todas as medições foram feitas em cm, crie uma nova coluna para cada atributo com seus valores em milímetros

```python
for column in iris_data.columns:
    if column.endswith(' cm'):
        new_column_name = column.replace(' cm', ' mm')
        iris_data[new_column_name] = iris_data[column] * 10
    
```

### 23. Calcule as medidas de media, variancia, obliquidade e kurtosi para cada um dos atributos.

```python
import scipy.stats as stats

atributos = data.iloc[:,:-1]
clasif = data['Classe']

saida = [["Classe", "Media", "Variancia", "Obliquidade", "Kurtosi"],]

for column in atributos.columns:
    media = atributos[column].mean()
    variancia = atributos[column].var()
    obliquidade = stats.skew(atributos[column])
    kurtosi = stats.kurtosis(atributos[column])

    
    saida.append([f"{column} {media} {variancia} {obliquidade} {kurtosi}"])
for line in saida:
    print("\t".join(line))

```
